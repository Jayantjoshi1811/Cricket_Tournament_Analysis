```python
import pandas as pd
df = pd.read_csv("ipl_summary.csv")
```

```python
df
```

```python
new_df = df.copy()
new_df['match_date'] = new_df['match_date'].fillna(method='ffill')
```

```python
new_df.shape
```

```python
new_df['match_title'] = new_df['match_title'].str.replace('PK','PBKS')
new_df
```

```python
df_balls = pd.read_csv("ipl_2025_deliveries.csv")
```

```python
df_balls
```

```python
df_balls['over_number'] = df_balls['over'].astype(str).str.split('.').str[0].astype(int)
```

```python
df_balls
```

```python
df_balls.iloc[15]
```

```python
match_summary = df_balls.groupby('match_id').agg({
    'runs_of_bat': 'sum',
    'extras': 'sum',
    'over': 'count'
}).reset_index()

match_summary.rename(columns={
    'runs_of_bat': 'total_runs',
    'over': 'total_balls',
} , inplace=True)
```

```python
match_summary
```

```python
runs_per_match = df_balls.groupby(['match_id', 'striker'])['runs_of_bat'].sum().reset_index()
top_scorer_per_match = runs_per_match.sort_values(['match_id', 'runs_of_bat'], ascending=[True, False])
top_scorer_per_match = top_scorer_per_match.groupby('match_id').head(1).reset_index(drop=True)
```

```python
top_scorer_per_match
```

```python
top_scorer_overall = df_balls.groupby('striker')['runs_of_bat'].sum().reset_index()
top_scorer_overall = top_scorer_overall.sort_values('runs_of_bat', ascending=False)
top_scorer_overall
```

```python
# Count valid dismissals per bowler per match
wickets_per_match = df_balls[df_balls['wicket_type'].notna()]
wickets_per_match = wickets_per_match.groupby(['match_id', 'bowler'])['wicket_type'].count().reset_index()

# Get top wicket-taker for each match
top_wicket_taker_per_match = wickets_per_match.sort_values(['match_id', 'wicket_type'], ascending=[True, False])
top_wicket_taker_per_match = top_wicket_taker_per_match.groupby('match_id').head(1).reset_index(drop=True)
```

```python
top_wicket_taker_per_match
```

```python
top_wicket_taker = df_balls.groupby('bowler')['wicket_type'].agg(lambda i: i.notna().sum()).reset_index().sort_values('wicket_type',ascending = False)
```

```python
top_wicket_taker[top_wicket_taker['wicket_type']=='0']
```

```python
top_scorer_overall = top_scorer_overall.reset_index()
```

```python
top_scorer_overall.rename(columns={'striker': 'player'}, inplace=True)
top_scorer_overall
```

```python
top_wicket_taker.rename(columns={'bowler': 'player'}, inplace=True)
top_wicket_taker
```

```python
player_stats = pd.merge(top_scorer_overall, top_wicket_taker, on='player' ,how='outer').fillna(0)
```

```python
player_stats
```

```python
player_stats['wicket_type'].max()
```

```python
player_stats.drop('index', axis =1, inplace =True)
player_stats
```

```python
final_match_summary = pd.merge(new_df, match_summary, on='match_id', how = 'left')
```

```python
final_match_summary
```

```python
final_match_summary.drop('total_runs', axis =1, inplace =True)
```

```python
final_match_summary.drop('total_balls', axis =1, inplace =True)
```

```python
final_match_summary
```

```python
# Total matches played by each team
team1= final_match_summary['team1_shortform'].value_counts()
team2= final_match_summary['team2_shortform'].value_counts()

total_matches = (team1 + team2).reset_index()
total_matches.columns = ['team', 'matches_played']

valid_teams = final_match_summary['team1_shortform'].unique().tolist() + final_match_summary['team2_shortform'].unique().tolist()
valid_teams = list(set(valid_teams))  # Remove duplicates

# Step 2: Count only valid winners
wins_df = final_match_summary[final_match_summary['Winner'].isin(valid_teams)]
wins = wins_df['Winner'].value_counts().reset_index()
wins.columns = ['team', 'wins']

# Merge total matches and wins
final_team_summary = pd.merge(total_matches, wins, on='team', how='left')

# Fill NaN values in wins 
final_team_summary['wins'] = final_team_summary['wins'].fillna(0).astype(int)

final_team_summary = final_team_summary.sort_values(by='wins', ascending=False).reset_index(drop=True)

```
